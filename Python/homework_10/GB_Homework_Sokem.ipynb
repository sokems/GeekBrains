f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30

1. Определить корни
2. Найти интервалы, на которых функция возрастает
3. Найти интервалы, на которых функция убывает
4. Построить график
5. Вычислить вершину
6. Определить промежутки, на котором f > 0
7. Определить промежутки, на котором f < 0

```python
import matplotlib.pyplot as plt
from math import *

def funct(x):
    co = cos(x)
    f = (-1) * 12 * x**4 * sin(co) - 18 * x**3 + 5 * x**2 + 10 * x - 30
    return f

x = list(range(1,100))
fx = [funct(i) for i in x]
plt.axes(facecolor='white')
plt.plot(x,fx)
plt.show()
```

Функция является переодичной, так как присутствует тригонометрические подфункции. Период равен числу Pi, следовательно интервал убывания и возрастания равен Pi/2. Вершину и провежутки f<0 и f>0 оперелить невозможно, так как это бесконечно расширяющаяся функция. Корни функции отображает график.
